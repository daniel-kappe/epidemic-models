# Die Idee

* Simulation der Ausbreitung einer ansteckenden Krankheit nach einem [SIR](https://de.wikipedia.org/wiki/SIR-Modell) ähnlichen Modell
* Die Bevölkerung, eine Fläche aus `N x M` Zellen, wird in drei Gruppen aufgeteilt:
  * Ansteckbare (Susceptible)
  * Angesteckte (Infected)
  * Genesene (Recovered)

1     |2  |...|N
------|---|---|---
N+1   |N+2|...|2N
...   |...|...|...
N(M-1)|...|...|NM

* Nachbarn stecken sich mit einer Wahrscheinlichkeit `p` an
* Eine Infektion dauert `d` Tage

## Frage
Wie verhält sich das Infektionsgeschehen für unterschiedliche Werte für `p` und `d`?

## Bonus
Wie verändert sich das Verhalten, wenn:
* genesene Personen nur `i` Tage lang immun sind?
* genesene Personen eine geringere Infektionswahrscheinlichkeit `q` haben?
* die einzelnen Personen unterschiedliche Anfälligkeit für die Krankheit haben?
